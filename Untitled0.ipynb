```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
```

```python
df = sns.load_dataset("titanic")
df.head()
```

#Bar chart

```python
df["sex"].value_counts().plot(kind='bar')
plt.show()
```

#Histogram

```python
plt.hist(df["age"])
plt.show()
```

# Marker "o"


```python
y = np.array([13,28,11,100])
plt.plot(y,marker= "o")
plt.show()
```

# Marker "*"

```python
plt.plot(y,marker= "*")
plt.show()
```

*** Common markers : ***
* "." : Point marker
* "," : Pixel marker
* "o" : Circle marker
* "v" : Triangle down marker
* "^" : Triangle up marker
* "<" : Triangle left marker
* ">" : Triangle right marker
* "1" : Tri-down marker
* "2" : Tri-up marker
* "3" : Tri-left marker
* "4" : Tri-right marker

# Marker Size
It call sometimes "markersize" or "ms". Adjust the size of the marker to make it more visible or match the importance of data points.

```python
plt.plot(y, marker="o", markersize=10)  # Adjust marker size
plt.show()
```

```python
## Marker Edge Color
change the color of the marker’s edge, which can help the markers stand out.
```

```python
plt.plot(y, marker="D", markeredgecolor="black", markersize=8)
plt.show()
```

## Marker Face Color
Set the fill color of the marker. This is especially useful when the edge color is different.

```python
plt.plot(y, marker="^", markerfacecolor="green", markeredgecolor="red", markersize=10)
plt.show()
```

##Custom Markers
You can create your own spesific symbols or character strings

```python
plt.plot(y, marker="$\u2665$", markersize=15)  # Heart symbol as marker
plt.show()
```

So let see all functions together

```python
plt.plot(y, marker="H", markersize=12, markerfacecolor="orange", markeredgecolor="black", markeredgewidth=1.5, alpha=0.7, linestyle="-.")
plt.show()
```

# Pairplot
Explore relationships between multiple numerical features.

```python
sns.pairplot(df, hue="survived")
plt.show()
```

## Violin Plot:
Compare age distributions by survival and passenger class.

```python
sns.violinplot(x="pclass", y="age", hue="survived", data=df, split=True)
plt.show()
```

## Swarm Plot:
Show all passengers' ages distributed by survival and gender.



```python
sns.swarmplot(x="sex", y="age", hue="survived", data=df)
plt.show()
```

##Joint Plot:
Compare age and fare with survival as a hue.

```python
sns.jointplot(x="age", y="fare", data=df, hue="survived", kind="scatter")
plt.show()
```

## FacetGrid:
Display multiple histograms of age by survival status.


```python
g = sns.FacetGrid(df, col="survived")
g.map(sns.histplot, "age")
plt.show()
```

## KDE Plot (Kernel Density Estimate):
Show a smoother distribution of ages by survival.



```python
sns.kdeplot(df.loc[df['survived'] == 1, "age"], fill=True, label="Survived")
sns.kdeplot(df.loc[df['survived'] == 0, "age"], fill=True, label="Did Not Survive")
plt.legend()
plt.show()
```

## Barplot with Multiple Categories:
Compare survival rates across passenger classes and gender.

```python
sns.barplot(x="pclass", y="survived", hue="sex", data=df)
plt.show()
```

## Bonus - Create your heart - shaped plot

```python
# Define the heart shape using a parametric equation
t = np.linspace(0, 2 * np.pi, 100)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

# Plot the heart shape
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red', linewidth=2)
plt.fill(x, y, color='pink', alpha=0.5)  # Fill the heart with color for a fun effect
plt.title("Heart-Shaped Plot")
plt.axis("equal")  # Make sure the heart isn't distorted
plt.axis("off")    # Turn off axis for a cleaner look
plt.show()
```
